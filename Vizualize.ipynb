```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

BAR PLOT

```python
# create data for bar plot

m = 30 # rows
n = 6 # column

# generate data
data = np.zeros((m, n))

for i in range(n):
  data[:, i] = 30*np.random.randn(m)*(2*i/(n-1)-1)**2 + (i+1)**2

```

```python
# show the bar
fig, ax = plt.subplots(1, 3, figsize = (8, 2))

# naked bar
ax[0].bar(range(n), np.mean(data, axis = 0))
ax[0].set_title('Bar plot')

# error plot
ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis = 0, ddof=1), marker = 's', linestyle = '')
ax[1].set_title('Errorbar plot')

# naked and error bar plot
ax[2].bar(range(n), np.mean(data, axis = 0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis = 0, ddof=1), marker = 's', linestyle = ' ', color = 'black')
ax[2].set_title('Bar plot and Errorbar plot')

plt.show()
```

```python
# amnually specify x axis

xcrossing = [1, 2, 4, 5, 6, 9]

plt.bar(xcrossing, np.mean(data, axis = 0))
plt.errorbar(xcrossing, np.mean(data, axis =0), np.std(data, axis = 0, ddof = 1), marker = 's', linestyle = ' ', color = 'black')
plt.title('Bar + Error plot')
plt.xticks(xcrossing)
plt.show()
```

```python
## note about bars from matrix

# data are into groups (rows) X properties (colunms)
m = [[2, 5, 4, 3], [1, 1, 8, 6]]
fig, ax = plt.subplots(2, 2, figsize = (8,8))

# conceputaulizin data as(rows) groups of (colunms)
ax[0,0].imshow(m)

# using pandas dataframe
df = pd.DataFrame(m, columns=['prop 0','prop 1', 'prop 2', 'prop 3'])
df.plot(ax = ax[1, 0], kind = 'bar')
ax[1, 0].set_title('Grouping by rows')

# now other orientation (property X group)
ax[0, 1].imshow(np.array(m).T)
df.T.plot(ax = ax[1, 1], kind = 'bar')
ax[1, 1].set_title('Grouping by column')

plt.show()
```

PIE CHART

```python
#create data for plot
Nbins = 5
totaln = 100
rawdata = np.ceil(np.logspace(np.log10(1/2), np.log10(Nbins-.01), totaln))

# create data for pie chart
uniquenu = np.unique(rawdata)
dataForPie = np.zeros(len(uniquenu))
```

```python
uniquenu
```

```python
for i in range(len(uniquenu)):
  dataForPie[i] = sum(rawdata==uniquenu[i])
```

```python
dataForPie
```

```python
plt.pie(dataForPie, labels = 100*dataForPie/sum(dataForPie))
plt.show()
```

```python
plt.pie(dataForPie,  labels = ['one', 'two', 'three', 'four', 'five'], explode=[0, .1, 0, .3, 0])
plt.show()
```

LINE PLOT

```python
n = 1000
data1 = np.exp(np.random.randn(n)/2)
data2 = np.exp(np.random.randn(n)/10)
data3 = np.exp(np.random.randn(n)/2 + 1)
k = 20

plt.hist(data1, bins=20)
plt.hist(data2, bins=20)
plt.hist(data3, bins=20)
plt.show()
```

```python
y1, x1 = np.histogram(data1, k)
y2, x2 = np.histogram(data2, k)
y3, x3 = np.histogram(data3, k)

xx1 = (x1[0: -1]+x1[1:]/2)
y1 = 100*y1/ sum(y1)
xx2 = (x2[0: -1]+x2[1:]/2)
y2 = 100*y2/ sum(y2)
xx3 = (x3[0: -1]+x3[1:]/2)
y3 = 100*y3/ sum(y3)

plt.plot(xx1, y1, 's-', label = 'data1')
plt.plot(xx2, y2, 'o-', label = 'data2')
plt.plot(xx3, y3, '^-', label = 'data3')
plt.legend()
plt.xlabel('Value')
plt.ylabel('probability')
plt.show()
```
